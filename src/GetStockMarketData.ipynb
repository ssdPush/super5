```python
import pandas as pd
import yfinance as yf #pip install yfinance
import datetime

# Define the stock symbols for the top 5 tech companies with their readable names
companies = {
    'AAPL': 'Apple',
    'MSFT': 'Microsoft',
    'AMZN': 'Amazon',
    'GOOGL': 'Google',
    'META': 'Meta'
}

# Define the start and end date for the data
start_date = datetime.datetime(2023, 1, 1)
end_date = datetime.datetime(2023, 12, 31)

# Fetch stock data for each company, ensure no ticker row is added, and add the `Company` column
data_frames = []
for symbol, name in companies.items():
    df = yf.download(symbol, start=start_date, end=end_date)
    df.columns = [col[0] if isinstance(col, tuple) else col for col in df.columns]
    df['Company'] = name  # Add the company name as a column
    data_frames.append(df)

# Concatenate all data into a single DataFrame
concatenated_data = pd.concat(data_frames)

# Save the concatenated data to a single CSV file
concatenated_data.to_csv("stockData.csv")
```
